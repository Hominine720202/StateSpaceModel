```python
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
from tqdm import tqdm_notebook
from scipy.stats import norm, poisson, gamma, uniform
sns.set(style="white")
```

```python
n_sample = 500
var1 = 1e-2
var2 = 5e-3
var2 = 1e-2
cov12 = 8e-3

mu = np.array([0, 0])
sigma = np.array([[var1, cov12], [cov12, var2]])

np.random.seed(1)
state_dist = np.random.multivariate_normal(mean=mu, cov=sigma, size=n_sample)
data = state_dist.cumsum(axis=0)

print("correlation coefficients: {0:.3f}".format(cov12 / np.sqrt(var1 * var2)))
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 6))
axes[0].plot(data[:,0])
axes[1].plot(data[:,1])
plt.tight_layout()

plt.subplots(figsize=(5, 4))
plt.scatter(state_dist[:,0], state_dist[:,1], s=5)
plt.tight_layout()


plt.subplots()
plt.scatter(data[:,0], data[:,1])

plt.subplots()
plt.plot(state_dist[:,0])
plt.plot(state_dist[:,0])
```
